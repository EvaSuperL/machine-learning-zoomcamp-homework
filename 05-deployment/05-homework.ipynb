````python
import requests
url = "http://127.0.0.1:8000/predict"
client = {
    "lead_source": "organic_search",
    "number_of_courses_viewed": 4,
    "annual_income": 80304.0
}
print(requests.post(url, json=client).json())
````

````python
!docker run --rm agrigorev/zoomcamp-model:2025 python -c "
import pickle

# 模型文件在根目录中
with open('pipeline_v2.bin', 'rb') as f:
    model = pickle.load(f)

client = {
    'lead_source': 'organic_search',
    'number_of_courses_viewed': 4,
    'annual_income': 80304.0
}

probability = model.predict_proba([client])[0][1]
print(f'Conversion probability: {probability:.3f}')

options = [0.39, 0.59, 0.79, 0.99]
closest = min(options, key=lambda x: abs(x - probability))
print(f'Closest option: {closest}')
print(f'Question 6 Answer: {closest}')
"
````
